```python
import pandas as pd
import pandas_datareader.data as web


df = web.DataReader(name='META', data_source='stooq')
df.head()
```

Zapis / Odczyt danych z pliku CSV

```python
df.to_csv('meta.csv')
```

```python
# Wycinanie danych dla konkretnego miesiąca
df_nov = df[(df.index.year == 2019) & (df.index.month == 11)]
df_nov
```

```python
df_nov.to_csv('meta_nov.csv')
```

```python
new_df = pd.read_csv('meta_nov.csv', index_col=0)
new_df
```

```python
df_nov.to_excel('meta_nov.xlsx')
new_df = pd.read_excel('meta_nov.xlsx', index_col = 0) #indeks_col = 0, bo chcemy, aby nadal kolumna z datą była naszym indeksem.
# Inaczej utworzona zostałaby dodatkowa kolumna z indeksem od 0
new_df
```
