```python
import numpy as np
import pandas as pd
pd.__version__
```

```python
url = ('https://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

```python
### DOBRĄ PRAKTYKĄ JEST ZOSTAWIĆ SOBIE SUROWE DANE, ABY MOŻNA BYŁO DO NICH WRÓCIĆ ###

# Kopiujemy nasze dane

retail = retail_raw.copy()

# Wyświetlamy informacje o tych danych

retail.info()
```

```python
retail.describe() #do odczytania statystyk / możemy odczytać np. ujemne błędne wartości do usunięcia itp
```

```python
# Sprawdzamy braki w danych

retail.isnull().sum()
```

```python
# Usuwamy braki w danych - usunięcte zostaną wszystkie wiersze z brakami

retail = retail.dropna()
```

```python
# Sprawdzamy jeszcze raz, czy braki zostały usunięte

retail.isnull().sum()
```

```python
# Co możemy zrobić z danymi, które są liczbami ujemnymi, a nie powinny być? Powinny być liczbą całkowitą 0 lub wyżej.
# Możemywyciąć dane z wartościami ujemnymi

retail.Quantity >= 0 #maska logiczna, którą wytniemy z naszego obiektu data frame

retail = retail[retail.Quantity >= 0] #otrzymujemy data frame z poprawnymi wartosciami i nadpisujemy naszą ramkę
```

```python
# Jeszcze raz sprawdzamy info

retail.info()
```

```python
# Konwertujemy CustomerID z float na wartość tekstową.
# Przypiszemy columnie customerid tę samą kolumnę z zastosowaną na niej "apply" i funkcją lambda. Najpierw konwersja na liczbą całkowitą - int(x),
# a potem int(x) na tekst str()

retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))

```

```python
# SELECT CustomerID, count(*) FROM retial GROUP BY CustomerID;
retail.groupby('CustomerID').size()
```

```python
# Do kolejnego zapytania potrzebujemy utworzyć nową kolumnę - Revenue (przychód) - quantity * unit price

retail['Revenue'] = retail.Quantity * retail.UnitPrice
```

```python
# SELECT CustomerID, avg(Revenue), count(*) FROM retial GROUP BY CustomerID;

retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size})

#To co zrobiliśmy wyżej to grupowanie po CustomerID, a potem podaliśmy funkcje agregujące mean i size przy uzyciu aggregate. W agregate przekazaliśmy słownik,
# który mapuje nazwy kolumn odpowiednio na funkcje, których chcieliśmy użyć.
```

```python
# Chcemy teraz zmienić nazwy powyższych kolumn wynikowych, bo te nazwy nie przedstawiają avg i count

retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size}). \
  rename(columns={'Revenue': 'RevenueAverage', 'CustomerID': 'CustomerIDCount'})
```

```python
# SELECT InvoiceDateDay, sum(Revenue) FROM retail GROUP BY InvoiceDateDay;

#Musimy stworzyć nową kolumnę z liczbą oznaczającą dany dzień miesiąca - dzień miesiąca można wyciąć dt.day

retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day
retail.head(2)
```

```python
results = retail.groupby('InvoiceDateDay').aggregate({'Revenue': np.sum})
```

```python
import seaborn as sns
sns.set()
results.plot(kind='bar')
```
