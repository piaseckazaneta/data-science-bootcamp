```python
import pandas as pd
import numpy as np

pd.__version__
```

```python
df = pd.read_csv('london_bike.csv')
df.head()
```

```python
df = df.set_index('timestamp') # ustawianie wybranej kolumny jako indeksu / wybieramy timestamp
df.head()
```

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

apple = fetch_financial_data('AAPL')
amazon = fetch_financial_data('AMZN')
google = fetch_financial_data('GOOGL')
uber = fetch_financial_data('UBER')
```

```python
apple.info()
```

```python
apple.head()
```

Preprocessing

```python
apple.columns = ['apple_' + col.lower() for col in apple.columns] #do każdej kolumny dodajemy przedrostek 'apple' i zamieniamy nazwy kolumn na małe litery / list comprahension
amazon.columns = ['amazon_' + col.lower() for col in amazon.columns]
google.columns = ['google_' + col.lower() for col in google.columns]
uber.columns = ['uber_' + col.lower() for col in uber.columns]
```

```python
apple.head()
```

Konkatenacja danych

```python
df = pd.concat(objs = [apple, amazon, google, uber], axis=1)
```

```python
df.describe()
```

```python
#Zmiana wyświetlania liczb z funckji wykłądniczej na 2 po przecinku

pd.set_option('display.float_format', lambda x: f'{x: .2f}')

df.describe().T
```

```python
df.corr() #liczy współczynniki korelacji poszczególnych zmiennych
```

```python
# Wyodrębnienie z df tylko tych kolumn z wartością zamknięcia 'close'

closes = [col for col in df.columns if col.endswith('close')]
closes
```

```python
# Policzenie korelacji dla closes
df[closes].corr()
```

```python
# Closes i volume

closes_vol = [col for col in df.columns if col.endswith('close') or col.endswith('volume')]
closes_vol
```

```python
df[closes_vol].corr()
```
