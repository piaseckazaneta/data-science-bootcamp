Podstawowe struktury danych - pd.DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = [12, 12, 32])
df
```

```python
df = pd.DataFrame(data = [12, 12, 32], index = ['first', 'second', 'third'], columns = ['col_1'])
df
```

```python
# DataFrame ze słownikiem
df = pd.DataFrame(data = {'WIG20': ['PKN ORLEN', 'PKO BP'],
                          'mWIG40': ['Amica', 'Playway']})
df
```

```python
df = pd.DataFrame(data =[[10, 12, 13], [23, 12, 10]], index = ['first', 'second'], columns = ['col_1', 'col_2', 'col_3'])
df
```

```python
# Zwróć nazwy kolumn
df.columns
```

```python
# Zwróć indeksy
df.index
```

```python
# Zwróć wartości - zwróci w ndarray numpy
df.values
```

```python
# Zwróć informację o obiekcie (non-null - nie ma wartości zerowych)
df.info()
```

```python
# Zwróć opis obiektu w postaci statystyk
df.describe()
```

Selekcja kolumn

```python
df
```

```python
df['col_1'] # ten zapis zwraca obiekt typu Series / w Series może być tylko 1 obiekt, jednego typu
```

```python
# Jeżeli chcemy wyciąć naszą kolumnę i otrzymać dataframe, musimy użyć podwójny nawias kwiadratowy [[]]
df[['col_1']]
```

Zmiana nazw kolumn

```python
df.columns = ['a', 'b', 'c'] #ten zapis nadpisze nam nazwy kolumn
df
```

```python
df.a #tak też można wyciąć kolumnę z dataframe, ale w wyniku dostajemy typ Series - trzeba o tym pamietac
```

```python
# Wygenerowanie nowej kolumny, zawierającej wartość sumaryczną z kolumn a i b
df['d'] = df.a + df.b
df
```

```python
# Inne metody
df.loc['first'] # metoda loc pozwala wycinać obiekty (wiersze) po nazwach
```

```python
df.iloc[0] # metoda loc wycinająca obiekty po numerach
```

```python
# Wycinanie pojedynczego elementu (trzeba podać który wiersz i która kolumna) - metodą loc
df.loc['first', 'b']
```

```python
df.loc[:,'a'] # wycinanie kolumny metodą loc (: - wszystkie wiersze, nazwa kolumny)
```
