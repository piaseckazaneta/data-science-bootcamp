```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

Metryki - Klasyfikacja binarna

Accuracy - dokładność klasyfikacji

Accuracy = correct predictions / total predictions * 100% *kursywa*



```python
y_true = np.array([0,1,1,0,0,0,0,1,1,0,1])
y_pred = np.array([0,1,1,0,0,1,1,0,1,0,1])
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_true, y_pred)
```

```python
# Stworzymy obiekt dataframe, bo TO JEST PODSTAWOWY TYP OBIEKTU DO TWORZENIA WYKRESÓW W PLOTLY

results = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred})
results
```

```python
# Robimy prostą wizualizację naszych predykcji
#Najpierw sortujemy y_true i przypisujemy do results, aby sortować na stałe

results = results.sort_values(by = 'y_true')
results = results.reset_index(drop=True) # po sortowaniu indeksy są wymieszane, więc reset
results['sample'] = results.index + 1 # tworzymy kolumne z nr próbki
results


```

Tworzymy wykres

```python
fig = make_subplots(rows = 2, cols=1) #tworzymy figure - będzie miało 2 wiersze, 1 kolumnę
fig.add_trace(go.Scatter(x=results['sample'], y=results['y_true'], mode = 'markers', name = 'y_true'), row=1, col=1) #tworzymy ślad wykresu 'add_trace'
# będzie to wykres punktowy 'go.Scatter'
fig.add_trace(go.Scatter(x=results['sample'], y=results['y_pred'], mode = 'markers', name = 'y_pred'), row=2, col=1)
fig.update_layout(width=900, height=500, title='Klasyfikator binarny') #aktualizujemy layout, podajemy wymiary
fig.show() #wyświetlamy

#Im bardziej dolny wykres różni się od górnego, tym model jest mniej dokładny
```
